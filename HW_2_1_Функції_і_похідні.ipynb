1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
f_lin = 10 * x - 5
df_lin = sp.diff(f_lin, x)
print("Похідна лінійної функції 10x-5:", df_lin)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 100)
f_lin = 10 * x - 5
df_lin = np.full_like(x, 10)
    
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_lin, label='f(x) = 10*x-5', color='blue')
plt.plot(x, df_lin, label="f'(x) = 10", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

Швидкість зміни оцінки постійна i дорівнює 10 балів за кожну годину навчання.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
from sympy import sin, ln
x = sp.symbols('x')

# sin(x)
f_sin = sin(x)
df_sin = sp.diff(f_sin, x)

# ln(x)
f_ln = ln(x)
df_ln = sp.diff(f_ln, x)

print("Похідна функції sin(x):", df_sin)
print("Похідна функції ln(x):", df_ln)
```

```python
# Визначення функції та її похідної
x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
f_sin = np.sin(x)
df_sin = np.cos(x)
    
# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x, df_sin, label="f'(x) = cos(x)", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10, 150)
x = x[x != 0] 
f_ln = np.log(x)
df_ln = 1/x

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x, df_ln, label="f'(x) = 1/x", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
func = x ** 2 - 5 * x + 1
dfunc = sp.diff(func, x)
print("Похідна функції x^2-5x+1:", dfunc)
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 100, 100)
func = x ** 2 - 5 * x + 1
dfunc = 2 * x - 5

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, func, label='f(x) = x^2 - 5x + 1', color='blue')
plt.plot(x, dfunc, label="f'(x) = 2*x - 5", color='green')

# Додавання легенди, міток та заголовка
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
